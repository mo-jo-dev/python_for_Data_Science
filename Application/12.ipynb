Loops
- For, while and do while
- Loops allow us to iterate
- Loops helps us iterate until a condition is specified

```python
seq = [1, 2, 3, 4, 5]
```

```python
for item in seq:  # The temporary variable name item can be whatever we want it to be
    print(item)   # Here, in for loop, the iteration happens based on what we write in the loop intro and not the body 
```

```python
i = 1
while i < 5:
    print(f'i is: {i}')
    i = i + 1    # Condition is written inside the body unlike the for loop
```

```python
for i in range(1, 11):
    print(f'12 x {i} = {12*i}')
```
