Dictionaries
- In key-value pair rather than indexing
- Similar to hash maps
- Mutable in nature

```python
d = {
    'key1' : 'value',
    'key2' : 123
}
```

```python
d['key1']
```

```python
d['key2']
```

```python
d = {
    'k1': [1, 2, 3] # With this type of case where list is shown as a value, we can use the normal indexing to access the elements
}
```

```python
d['k1']
```

```python
d['k1'][1]
```

```python
my_list = d['k1']
```

```python
my_list[1]    
```

```python
d = {
    'k1': {
        'name':'k1 buddy',
        'class': 2,
        'place': 'nothing',
        'marks': [1, 2, 3, 4]
    }
}
```

```python
d['k1']
```

```python
d['k1']['marks'][2]
```

```python
# Dictionaries do not retain any order, they are just carrying key value pairs
```
